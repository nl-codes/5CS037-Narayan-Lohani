# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:
  
What percent of the total fare revenue came from each passenger class?  
To answer the question perform following operation:
1. Calculate the total Fare paid across all passengers in the Titanic dataset.
2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.
3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).
4. Combine the fare totals from first, second, and third classes into a list.
5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```python
total_fare = titanic_dataset['Fare'].sum()

first_class_fare = titanic_dataset[titanic_dataset['Pclass'] == 1]['Fare'].sum()
second_class_fare = titanic_dataset[titanic_dataset['Pclass'] == 2]['Fare'].sum()
third_class_fare = titanic_dataset[titanic_dataset['Pclass'] == 3]['Fare'].sum()

class_fare_list = [first_class_fare, second_class_fare, third_class_fare]

fare_by_class = {}
for i in range(len(class_fare_list)):
  fare_by_class[f'Class {i+1}'] = class_fare_list[i]/total_fare
print(f"Result using instructions : {fare_by_class}")

# Solving using Group-by method:
fare_by_class_group = titanic_dataset.groupby('Pclass')['Fare'].sum()
fare_proportions = fare_by_class_group/total_fare
print(f"Result using Group-by method : \n{fare_proportions}")
```

What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?  
To answer the question perform following operation:
1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18-64), and "senior" (age 65 and above).
2. Calculate the total number of passengers on the Titanic.
3. Count the number of passengers in each age group.
4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.
5. Display the proportion as a percentage.

```python
def categorize_age(age):
  """
  Function to categorize age into child, adult and senior
  Parameters:
  age(int) : age of the person
  Returns:
  string : category of the person (child or adult or senior)
  """
  if age < 18:
      return 'child'
  elif 18 <= age < 65:
      return 'adult'
  else:
      return 'senior'

titanic_dataset['age_group'] = titanic_dataset['Age'].apply(categorize_age)
total_passengers = len(titanic_dataset)

child_count = len(titanic_dataset[titanic_dataset['age_group'] == 'child'])
adult_count = len(titanic_dataset[titanic_dataset['age_group'] == 'adult'])
senior_count = len(titanic_dataset[titanic_dataset['age_group'] == 'senior'])

count_by_age_group = [child_count, adult_count, senior_count]
passenger_category = ['child', 'adult', 'senior']

age_group_proportions = {}
for i in range(len(count_by_age_group)):
  age_group_proportions[f'Age Group {passenger_category[i]}'] = count_by_age_group[i]/total_passengers

print(f"Result using instructions : {age_group_proportions}")

# Solving using Group-by method
passenger_by_age_group = titanic_dataset.groupby('age_group').size()
age_group_proportions = passenger_by_age_group/total_passengers
print(f"Result using Group-by method : \n{age_group_proportions}")
```

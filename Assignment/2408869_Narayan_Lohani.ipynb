Except for India and Pakistan, all other rankings do align. Calculating to what degree the ranks align using Karl Pearson's Correlation Coefficient.

```python
def pearson_correlation(x, y):
    """
    Calculate the Pearson correlation coefficient between two variables.

    The Pearson correlation coefficient measures the linear relationship between two variables.
    It ranges from -1 to 1, where:
    -> 1 indicates a perfect positive linear relationship,
    ->-1 indicates a perfect negative linear relationship,
    -> 0 indicates no linear relationship for the given range of inputs

    Args:
        x (list or numpy array): A list or array like values for the first variable.
        y (list or numpy array): A list or array like values for the second variable.

    Returns:
        float: The Pearson correlation coefficient between x and y.

    Raises:
        ValueError: If the lengths of x and y do not match or if either is empty.
    """
    # Ensure x and y are numpy arrays for element-wise operations
    x = np.array(x)
    y = np.array(y)

    # Check if lengths match
    if len(x) != len(y):
        raise ValueError("Input lists must have the same length.")

    # Check if either array is empty
    if len(x) == 0 or len(y) == 0:
        raise ValueError("Input lists must not be empty.")

    # Calculate the Pearson correlation coefficient
    numerator = np.sum((x - np.mean(x)) * (y - np.mean(y)))
    denominator = np.sqrt(np.sum((x - np.mean(x))**2) * np.sum((y - np.mean(y))**2))

    return numerator / denominator
```

```python
# Calculating correlation coefficient between score and composite score
corr_coefficient_for_ranks = pearson_correlation(sa_rank_score['score'], sa_rank_composite_score['Composite_Score'])

print(f"Pearson's Correlation Coefficient for measure of linear realtionship between Rank based on ordinary score and composite score = {corr_coefficient_for_ranks}")
```

Pearson's Correlation Coefficient suggest that rank by ordinary score and rank by composite score have very strong positive linear relationship. Thus, they align very strongly.

# **Task- 3- Outlier Detection:**

 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
 2. Define outliers using the 1.5 × IQR rule.
 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
 4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
# Defining outliers using 1.5 IQR Rule
def detect_iqr_outliers(data, column):
  """
  Detects outliers using the IQR method.
  IQR = Interquartile Range = Q3 - Q1
  OUTLIERS = (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)

  Parameters:
  - data: DataFrame containing the data.
  - column: Column name to analyze.

  Returns:
  - A boolean Series where True indicates an outlier.
  """
  q1 = data[column].quantile(0.25)
  q3 = data[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr

  return (data[column] < lower_bound) | (data[column] > upper_bound)
```

```python
# Listing columns to find outliers in dataset
outlier_columns = ['Log GDP per capita', 'score']

# Adding column_name_outlier column and computing value using
# detect_iqr_outlier for elements in outlier_columns
for col in outlier_columns:
    sa_dataset[f'{col}_Outlier'] = detect_iqr_outliers(sa_dataset, col)

# Combine outliers into single column using bolean OR provided by .any() method
sa_dataset['Outlier'] = sa_dataset[[f'{col}_Outlier' for col in outlier_columns]].any(axis=1)

# Split the dataset into outliers and non-outliers
non_outliers = sa_dataset[~sa_dataset['Outlier']]
outliers = sa_dataset[sa_dataset['Outlier']]


# Create the scatter plot
plt.figure(figsize=(12, 6))

# Plot non-outliers
plt.scatter(
    non_outliers['Log GDP per capita'],
    non_outliers['score'],
    s=100,
    color='blue',
    edgecolor='black',
    alpha=0.7,
    label='Non-Outliers'
)

# Plot outliers
plt.scatter(
    outliers['Log GDP per capita'],
    outliers['score'],
    s=100,
    color='red',
    edgecolor='black',
    alpha=0.7,
    label='Outliers'
)

# Decorate the plot
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Addding labels, and titles
plt.xlabel('Log GDP per capita', fontsize=14, fontweight='bold')
plt.ylabel('Happiness score', fontsize=14, fontweight='bold')
plt.title('Scatter Plot : Log GDP per capita vs Happiness score with Outliers Highlighted Among South Asian Countries', fontsize=16, fontweight='bold')

# Adding grid to the plot
plt.grid(linestyle='--', alpha=0.7)

# Adding legend to the plot
plt.legend(fontsize=12)

# Adding country name in top of outliers
for gdp, score, country_name in zip(outliers['Log GDP per capita'], outliers['score'], outliers['Country name']):
    plt.text(x=gdp+0.01, y=score+0.01, s=country_name, fontsize=12, color='red')

# Optimize the plot
plt.tight_layout()

# Show the plot
plt.show()
```
